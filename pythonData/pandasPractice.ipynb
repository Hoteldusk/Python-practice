9. 0-100 사이 점수를 5개 연속으로 입력한 뒤 그 중 최고 점수를 찾아 출력하는 코드를 작성한다.

```python
# 0에서 100 사이의 점수를 5개 입력받아 최고 점수를 찾는 함수
def find_highest_score():
    scores = []
    for i in range(1, 6):
        while True:
            score = int(input(f"{i}번째 점수를 입력하세요 (0~100): "))
            if 0 <= score <= 100:
                break
            else:
                print("입력값이 범위를 벗어났습니다. 다시 입력해주세요.")
        scores.append(score)
    
    highest_score = max(scores)
    return highest_score

result = find_highest_score()

print("최고 점수:", result)
```

33. 북구 아파트 평당거래금액의 월별 변화를 선형 그래프로 작성한다. 그래프제목, x축, y축 제목, 범례 적절히 입력.
- 단, 문흥동, 일곡동, 운암동, 용봉동, 북구 평균을 색과 마커가 다른 5개의 선 그래프로 겹쳐서 작성한다.

```python
filtered_data = df2[df2['구'] == '북구']
filtered_data = filtered_data[(filtered_data['동'] == '문흥동') | (filtered_data['동'] == '일곡동') | (filtered_data['동'] == '운암동') | (filtered_data['동'] == '용봉동')]

average_price_per_month = filtered_data.groupby(['계약년월', '동'])['평당거래가격'].mean().unstack()

colors = ['red', 'green', 'blue', 'orange', 'purple']
markers = ['o', 's', '^', 'D', 'v']

for i, dong in enumerate(average_price_per_month.columns):
    plt.plot(average_price_per_month.index, average_price_per_month[dong], color=colors[i], marker=markers[i], label=dong)

plt.title("북구 아파트 평당거래금액의 월별 변화")
plt.xlabel("월")
plt.ylabel("평당거래금액")

plt.legend()

plt.show()
```

35. 아파트 평당거래금액과 거래건수의 월별 변화를 선형 그래프와 막대그래프로 중첩하여 작성한다. 그래프제목, x축, y축 제목, 범례 적절히 입력.
- 월별(1-12) 평균 평당거래금액을 선형 그래프로 작성
- 월별 거래건수는 막대그래프로 표현

```python
monthly_avg_price_per_area = df2.groupby('계약년월')['평당거래가격'].mean()
monthly_transaction_count = df2.groupby('계약년월').size()


plt.plot(monthly_avg_price_per_area.index, monthly_avg_price_per_area, marker='o', label='평당거래금액')
plt.xlabel('월')
plt.ylabel('평당거래금액')
plt.title('아파트 평당거래금액 및 거래건수 월별 변화')

plt.legend()

# 막대그래프 그리기
plt.bar(monthly_transaction_count.index, monthly_transaction_count, alpha=0.5, label='거래건수')
plt.xlabel('월')
plt.ylabel('거래건수')

plt.legend()

plt.show()
```
